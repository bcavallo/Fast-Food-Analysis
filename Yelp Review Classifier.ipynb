## Yelp Review Classification ##

In this ipython notebook I will detail the construction of the classifer that I use to study/visualize the sentiment of tweets in real time on http://45.55.185.83:5000/.

As a training set I used Yelp's corpus of reviews from 10 diverse geographic areas (see http://www.yelp.com/dataset_challenge).  

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

We first load our tables of all yelp reviews and all business reviews.  Using business descriptions from the business reviews, we then select all restaurant reviews and finally isolate our training set, all restaurant reviews with 1 or 5 stars.

```python
reviews = pd.read_csv('/home/vagrant/yelp_dataset_challenge_academic_' +  
                      'dataset/yelp_academic_dataset_review.csv')

biz = pd.read_csv('/home/vagrant/yelp_dataset_challenge_academic_dataset/yelp_academic_dataset_business.csv')

reviews = reviews.merge(biz[['business_id', 'categories']], on='business_id', how='left')

rest_reviews = reviews[reviews['categories'].str.contains('Restaurant')]

non_neutral = rest_reviews[rest_reviews.stars.isin([1,5])]

#tagging if a review is positive or negative
def good_bad(n):
    '''
    Classifies a 1 or 5 star review as negative or positive.
    
    @param n, A star rating
    returns 0 for 1 star, 1 for 5 stars
    '''
    
    if n == 1:
        return 0
    else:
        return 1
```

```python
non_neutral['target'] = non_neutral.stars.apply(good_bad)
```

Here is the proportion of positive to negative reviews.  Almost all reviews are positive.

```python
non_neutral.target.value_counts() / len(non_neutral)
```

We train a Multinomial Naive Bayes classifier to detect positive vs. negative reviews.  As these reviews are truly positive or truly negative, the classifier gets a better idea of more pointed language.  To train the classifier we use the bag-of-words representation of the reviews.  We also only consider words that occur at least 25 times and remove all english stop words.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english', min_df=25, ngram_range=(1,2))
```

```python
X = cv.fit_transform(non_neutral.text)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
```

```python
from sklearn.cross_validation import KFold, cross_val_score, train_test_split
```

The classifier performs well after a 10-fold cross-validation on the training set.

```python
cross_val_score(nb, X, non_neutral.target, cv=10)
```

```python
nb.fit(X, non_neutral.target)
```

```python
nb.score(X, non_neutral.target)
```

If we view the classifier as a linear model, we can find the coefficient corresponding to each feature.  Viewing these coefficients as weights we can find some of the most important positive and negative features.

```python
coef = nb.coef_.ravel()
positive_coefficients = np.argsort(coef)[-50:]
negative_coefficients = np.argsort(coef)[:50]
feature_names = np.array(cv.get_feature_names())
```

```python
plt.figure(figsize=(12,8))
plt.subplot(1,1,1)
plt.title("Importance of positive words")
plt.bar(np.arange(50),-1 * coef[positive_coefficients], color="blue")
plt.xticks(np.arange(1, 51), feature_names[positive_coefficients], rotation=60, ha="right")
plt.show()
plt.figure(figsize=(12,8))
plt.subplot(2,1,1)
plt.title("Importance of negative words")
plt.bar(np.arange(50), -1 * coef[negative_coefficients], color="red")
plt.xticks(np.arange(1, 51), feature_names[negative_coefficients], rotation=60, ha="right")
plt.show()
```

Now, let's use this classifier to predict the star rating of a restaurant based on all of its reviews.  To do this we first calculate the actual star rating as the average of all reviews.  We then compare it to our predicted star rating which we obtain by viewing each positive classification as a 5 star review, each negative classification as a 1 star review, and then taking the average of all reviews.  We restrict to restaurants with over 50 reviews.

```python
calc_avg = rest_reviews.merge(biz[['business_id', 'review_count']], on='business_id', how='left')
```

```python
calc_avg = calc_avg[calc_avg['review_count'] > 50]
```

```python
len(calc_avg)
```

```python
calc_avg = calc_avg.dropna(subset=['text'])
```

```python
C = cv.transform(calc_avg.text)
```

```python
preds = nb.predict(C)
```

```python
def to_revs(n):
    if n == 0:
        return 1
    if n == 1:
        return 5
```

```python
calc_avg['pred'] = map(to_revs, preds)
```

```python
guessed_revs = pd.DataFrame(calc_avg.groupby(['business_id']).pred.mean())
```

```python
actual_revs = pd.DataFrame(calc_avg.groupby(['business_id']).stars.mean())
```

```python
short = biz[['business_id', 'stars']]
```

```python
results = short.merge(guessed_revs[['pred']], left_on='business_id', right_index=True, how='right')
```

```python
results = results.merge(actual_revs[['stars']], left_on='business_id', right_index=True, how='right')
```

```python
np.abs((results['stars_y'] - results['pred'])**2).mean()
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
fit = np.polyfit(results['pred'], results['stars_y'], 1)
fit_fn = np.poly1d(fit)

plt.figure(figsize=(20,10))
plt.scatter(results['pred'], results['stars_y'], alpha=0.5)
plt.plot(np.arange(7), fit_fn(range(7)), color='#53F22C',alpha=.9, linewidth=3)
plt.axis([0,5.5,0,5.5])
plt.xlabel('Predicted Rating', fontsize=22)
plt.ylabel('Actual Rating', fontsize=22)
plt.title('Predicted Yelp Rating vs. Actual Rating', fontsize=22)
plt.show()
```
