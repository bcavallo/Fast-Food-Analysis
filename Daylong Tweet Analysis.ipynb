For this demonstration, I'll focus on the chains McDonald's and Starbucks.  Let's begin by plotting the total number of tweets corresponding to each restaurant in addition to the distribution of tweets for each restaurant over time.  To search for individual restaurants we will first turn all texts into lower case in addition to removing some key punctuation and other parts of tweets that will not contribute to our analysis.  For the distribution of tweets we'll also sort all times into 30 minute buckets.  In this case each bar represents the number of tweets from 10 minutes prior to the listed time up to the listed time.

```python
def clean_tweet(tweet):
    '''
    Returns a cleaned lowercase version of the tweet to help with searching.
    
    @param tweet, The text of a tweet
    @return A version of the tweet with rt, links, and apostrophes removed.
    '''
    res = re.sub(r'\shttps?:(.+)','', tweet.lower()) # Removes all links
    res = re.sub("^rt\s@.+?:\s","",res) # Removes the the "retweet" tag
    return re.sub("[^\w\s]", "", res) # Removes all punctuation

def adj_time(t):
    '''
    Returns a parsed version of unix timestamp rounded up to the nearest half hour mark (e.g. 8:05
    would be rounded up to 8:30)
    
    @param t, a unix timestamp
    @return A datetime corresponding to t rounded up to the nearest 30 minute mark
    '''
    adj = int(t) - (int(t) % 1800) + 1800
    return datetime.datetime.fromtimestamp(int(adj)).strftime('%m-%d %H:%M')


data['text_clean'] = data.text.apply(clean_tweet)
data['adj_time'] = data.time.apply(adj_time)
```

## Sentiment Analysis of Tweets

We now discuss the sentiment analysis of tweets from these restaurants.  Throughout the day we compute an overall sentiment score for each restaurant as captured by the tweets from the previous hour.  Every ten minutes we calculate the sentiment index as the number of positive tweets from the last hour divided by the total number of tweets from the last hour.

To classify the tweets, we use a Naive Bayes classifier trained on Yelp reviews from (LINK).  We first re-adjust the time so that we use 10 minute bins.  Following that we classifiy all of our tweets and plot the sentiment index time series for each restaurant.  By investigating sudden changes in the overall sentiment, it is possible to detect events pertaining to the restaurant.

```python
def re_adj_time(t):
    '''
    Returns a parsed version of unix timestamp rounded up to the nearest 10 minute
    mark (e.g. 8:05 would be rounded up to 8:10)
    
    @param t, a unix timestamp
    @return A datetime corresponding to t rounded up to the nearest 30 minute mark
    '''
    adj = int(t) - (int(t) % 600) + 600
    return datetime.datetime.fromtimestamp(int(adj)).strftime('%m-%d %H:%M')

mc_tweets['adj_time'] = mc_tweets.time.apply(re_adj_time)
star_tweets['adj_time'] = star_tweets.time.apply(re_adj_time)
```

```python
# Uploading classifier and count vectorizer

cv = joblib.load('/home/vagrant/yelp_dataset_challenge_academic_dataset/get_tweets/classifiers/cv.pkl')
nb = joblib.load('/home/vagrant/yelp_dataset_challenge_academic_dataset/get_tweets/classifiers/nb.pkl')
```

```python
# To help with classification we create a set of stopwords and also
# a function that removes stop words

my_stop_words = stopwords.words('english') + [u'mcdonalds',u'starbucks',u'kfc'] # This uses NLTK's list of stop words

def rem_stop_words(s):
    '''
    Removes all stop words from a given string.
    
    @param s, A string
    @return The input string with all stopwords removed.
    '''
    return ' '.join([word for word in s.split() if word not in my_stop_words])
```

```python
# This function computes running averages so that we can find the sentiment index
# of a given restaurant.

def running_avg(arr1, arr2, N):
    '''
    Calculates a running average where arr1 is the values, arr2 is the counts,
    and N is the span of the average.
    
    @param arr1, a column in a dataframe corresponding to the values we are averaging
    @param arr2, a column in a dataframe corresponding to the counts of the values we are averaging
    @param N, the span of the running average
    
    return A numpy array of the average after each step.
    '''
    i = 0
    size = len(arr1)
    res = []
    while i + N < size:
        res.append(float(arr1.values[i:i+N].sum()) / arr2.values[i:i+N].sum())
        i += 1
    return res
```

```python
# I now load the transform the tweets into their bag-of-words representation
# and calculate positivity vs. negativity according to the classifier.

start = time.time()
for i in [mc_tweets, star_tweets]:
    X = cv.transform(i.text_clean.apply(rem_stop_words))
    i['sent'] = nb.predict(X)

print "The above computation took %.02f seconds" % (time.time() - start)
```

```python
#ADD A LEGEND HERE TOMORROW!!!!!!!!!!!!!!!

mc_avgs = running_avg(mc_tweets.groupby(['adj_time']).sent.sum(), mc_tweets.groupby(['adj_time']).sent.count(), 6)
st_avgs = running_avg(star_tweets.groupby(['adj_time']).sent.sum(), star_tweets.groupby(['adj_time']).sent.count(), 6)

mc_ts = pd.Series(mc_avgs, index=mc_tweets.adj_time.unique()[6:])
st_ts = pd.Series(st_avgs, index=star_tweets.adj_time.unique()[6:])

plt.figure(figsize=(20, 10))
mc_ts.plot()
st_ts.plot()
plt.yticks
plt.ylabel("Sentiment Index")
plt.title("Sentiment Index over a 24 period")
plot.show()
```

I will begin by discussing the sharp rise and then fall that the McDonald's sentiment index time series experiences from roughly 12:40 to 15:30.  To do this we isolate the the tweets from that part of the time series and then perform Latent Dirichlet Allocation to extract topics.  Finally we then utilize the topic words to investigate the tweets more closely find the events that are triggering them.

Let's first study the rise of the time series.  This occurs from around 12:40 to 13:50.

```python
beg = 1427733600  # Timestamp for 12:40
end = 1427737800  # Timestamp for 13:50

mc_rise = mc_tweets[(mc_tweets.time > beg) & (mc_tweets.time < end)]

print "Total number of tweets in mc_rise is: %d" % len(mc_rise)
```

```python
cv2 = CountVectorizer(min_df = 10, stop_words=my_stop_words)

model = lda.LDA(n_topics=4, n_iter=1500, random_state=1)

Y = cv2.fit_transform(mc_rise.text_clean.apply(rem_stop_words))

model.fit(Y)
```

```python
topic_word = model.components_
n_top_words = 5
vocab = cv2.get_feature_names()

for i, topic_dist in enumerate(topic_word):
    topic_words = np.array(vocab)[np.argsort(topic_dist)][:-n_top_words:-1]
    print('Topic {}: {}'.format(i, ' '.join(topic_words)))

```
